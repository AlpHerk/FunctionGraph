```python
from matplotlib import pyplot as plt
from handcalcs import render
import sympy as sym
import numpy as np
```

```python
x = np.arange(-10, 10, 0.1)
plt.figure(figsize=(20, 10))

shx = np.sinh(x)
plt.subplot(2, 4, 1)
plt.title("双曲正弦")
plt.plot(x, shx)

chx = np.cosh(x)
plt.subplot(2, 4, 2)
plt.title("双曲余弦")
plt.plot(x, chx)

thx = np.tanh(x)
plt.subplot(2, 4, 3)
plt.title("双曲正切")
plt.plot(x, thx)

arshx = np.arcsinh(x)
plt.subplot(2, 4, 5)
plt.title("反双曲正弦")
plt.plot(x, arshx)

archx = np.arccosh(x)
plt.subplot(2, 4, 6)
plt.title("反双曲余弦")
plt.plot(x, archx)

arthx = np.arctanh(x)
plt.subplot(2, 4, 7)
plt.title("反双曲正切")
plt.plot(x, arthx)

plt.subplot(1, 4, 4)
plt.plot(x, shx)
plt.plot(x, chx)
plt.plot(x, thx)
plt.plot(x, arshx)
plt.plot(x, archx)
plt.plot(x, arthx)



# plt.subplots_adjust(wspace=1, hspace=1)
plt.show()
```
